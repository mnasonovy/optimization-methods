```python
import numpy as np
import matplotlib.pyplot as plt

# Функция
def f(x):
    return 1 - np.exp(-(x - 2)**2)

# Интервал и значения
x = np.linspace(-3, 5, 500)
y = f(x)

# График
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r"$f(x) = 1 - e^{-(x - 2)^2}$", color="navy")
plt.axvline(x=2, color="red", linestyle="--", label=r"Минимум при $x = 2$")
plt.scatter([2], [f(2)], color="red", zorder=5)

plt.title(r"График функции $f(x) = 1 - e^{-(x - 2)^2}$", fontsize=14)
plt.xlabel(r"$x$", fontsize=12)
plt.ylabel(r"$f(x)$", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Первая и вторая производные
def df(x):
    return 2 * (x - 2) * np.exp(-(x - 2)**2)

def d2f(x):
    return 2 * np.exp(-(x - 2)**2) * (1 - 2 * (x - 2)**2)

x = np.linspace(-3, 5, 500)

# f'(x)
plt.figure(figsize=(7, 4))
plt.plot(x, df(x), label=r"$f'(x)$", color="green")
plt.axhline(0, color="gray", linestyle=":")
plt.axvline(2, color="red", linestyle="--", label=r"$x = 2$")
plt.title("Первая производная $f'(x)$", fontsize=12)
plt.xlabel("$x$")
plt.ylabel("$f'(x)$")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()

# f''(x)
plt.figure(figsize=(7, 4))
plt.plot(x, d2f(x), label=r"$f''(x)$", color="purple")
plt.axhline(0, color="gray", linestyle=":")
plt.axvline(2, color="red", linestyle="--", label=r"$x = 2$")
plt.title("Вторая производная $f''(x)$ — выпуклость и вогнутость", fontsize=12)
plt.xlabel("$x$")
plt.ylabel("$f''(x)$")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определение функции
def f(x):
    return 1 - np.exp(-(x - 2)**2)

x = np.linspace(1.5, 2.5, 400)
y = f(x)
approx = (x - 2)**2

# График
plt.figure(figsize=(7, 4))
plt.plot(x, y, label=r"$f(x) = 1 - e^{-(x - 2)^2}$", color="navy")
plt.plot(x, approx, '--', label=r"Аппроксимация: $(x - 2)^2$", color="orange")
plt.title("Сравнение $f(x)$ и $(x - 2)^2$ в окрестности $x = 2$", fontsize=13)
plt.xlabel("$x$", fontsize=12)
plt.ylabel("$f(x)$", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 1 - np.exp(-(x - 2)**2)

# история сужения интервалов (N0–N12)
history = [
    (-3.000000, 5.000000),   # N0
    (0.999995, 5.000000),    # N2
    (0.999995, 3.000002),    # N4
    (1.999994, 3.000002),    # N6
    (1.999994, 2.500003),    # N8
    (1.999994, 2.250003),    # N10
    (1.999994, 2.125004)     # N12
]

x = np.linspace(-3, 5, 400)
y = f(x)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y, color='lightsteelblue', lw=2, label=r'$f(x) = 1 - e^{-(x - 2)^2}$')
ax.axvline(2, color='red', linestyle='--', lw=1.2, label='Минимум x = 2')

# палитра — разные мягкие оттенки
colors = ['#7aa2f7', '#9d7cd8', '#bb9af7', '#e0af68', '#e06c75', '#98c379', '#56b6c2']

# строим скобы
for i, (a_i, b_i) in enumerate(history):
    color = colors[i % len(colors)]
    y_bottom = -0.10 - 0.05 * i  # -0.10, -0.15, -0.20, ...
    # вертикальные линии до графика
    ax.plot([a_i, a_i], [y_bottom, f(a_i)], color=color, lw=1.6)
    ax.plot([b_i, b_i], [y_bottom, f(b_i)], color=color, lw=1.6)
    # нижняя перекладина
    ax.plot([a_i, b_i], [y_bottom, y_bottom], color=color, lw=1.6, label=f'Интервал N{i*2}')

# оформление
ax.axhline(0, color='black', lw=1.3)  # ось OX
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_ylim(-0.5, 1.05)
ax.set_xlim(-3.5, 5.5)
ax.set_title('Сужение интервала (итерации N0–N12)', fontsize=13)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=9)
ax.grid(True, alpha=0.25)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 1 - np.exp(-(x - 2)**2)

# история сужения интервалов (N14–N20)
history_14_20 = [
    (1.999994, 2.125004),   # N14
    (1.999994, 2.062504),   # N16
    (1.999994, 2.031254),   # N18
    (1.999994, 2.015629)    # N20
]

# диапазон по X — ориентир по границам интервалов
x_left, x_right = 1.995, 2.14
x = np.linspace(x_left, x_right, 800)
y = f(x)

fig, ax = plt.subplots(figsize=(9.5, 5))
ax.plot(x, y, color='lightsteelblue', lw=2, label=r'$f(x) = 1 - e^{-(x - 2)^2}$')
ax.axvline(2, color='red', linestyle='--', lw=1.2, label='Минимум x = 2')

# цвета
colors = ['#7aa2f7', '#9d7cd8', '#bb9af7', '#e0af68']

# строим аккуратные скобы
for i, (a_i, b_i) in enumerate(history_14_20):
    color = colors[i % len(colors)]
    y_bottom = -0.0009 * (i + 1)  # компактное расстояние между уровнями
    ax.plot([a_i, a_i], [y_bottom, f(a_i)], color=color, lw=1.4)
    ax.plot([b_i, b_i], [y_bottom, f(b_i)], color=color, lw=1.4)
    ax.plot([a_i, b_i], [y_bottom, y_bottom], color=color, lw=1.4, label=f'Интервал N{14 + 2*i}')

# оформление
ax.axhline(0, color='black', lw=1.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# границы по осям
ax.set_xlim(x_left, x_right)
y_top = y.max() + 0.002
y_bottom = min(-0.0009 * (len(history_14_20) + 1), -0.006)
ax.set_ylim(y_bottom, y_top)

ax.set_title('Сужение интервала (итерации N14–N20)', fontsize=13)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=9)
ax.grid(True, alpha=0.25)
plt.tight_layout()
plt.show()
```
